```python
import matplotlib.pyplot as plt
import csv
import numpy as np
def plot_curve(element):
    CASSCF_FILENAME = './casscf/{0}2_PES.csv'.format(element)
    DFT_FILENAME = './dft/{0}2_DFT_PES.csv'.format(element)
    MCPDFT_FILENAME = './mcpdft/{0}2_mcpdft_PES.csv'.format(element)
    with open(CASSCF_FILENAME, 'r') as f:
        reader = csv.reader(f)
        r_casscf = np.array(next(reader), dtype=float)
        e_casscf = np.array(next(reader), dtype=float)
        
    with open(DFT_FILENAME, 'r') as f:
        reader = csv.reader(f)
        r_DFT = np.array(next(reader), dtype=float)
        e_DFT = np.array(next(reader), dtype=float)

    with open(MCPDFT_FILENAME, 'r') as f:
        reader = csv.reader(f)
        r_MCPDFT = np.array(next(reader), dtype=float)
        e_MCPDFT = np.array(next(reader), dtype=float)

    # Convert eH to eV
    e_casscf *= 27.2114
    e_DFT *= 27.2114
    e_MCPDFT *= 27.2114
    # plt.plot(r_casscf, e_casscf, 'k', label='CASSCF ($\Delta E_d$={0:.3f}@{1:.2f}$\AA$)'.format(min(e_casscf)-e_casscf[-1], r_casscf[np.argmin(e_casscf)]))
    # plt.plot(r_DFT, e_DFT, 'r', label='B3LYP ($\Delta E_d$={0:.3f}@{1:.2f}$\AA$)'.format(min(e_DFT)-e_DFT[-1], r_DFT[np.argmin(e_DFT)]))

    plt.plot(r_casscf, e_casscf-min(e_casscf), 'k', label='CASSCF ($\Delta E_d$={0:.3f}@{1:.2f}$\AA$)'.format(e_casscf[-1]-min(e_casscf), r_casscf[np.argmin(e_casscf)]))
    plt.plot(r_DFT, e_DFT-min(e_DFT), 'r', label='B3LYP ($\Delta E_d$={0:.3f}@{1:.2f}$\AA$)'.format(e_DFT[-1]-min(e_DFT), r_DFT[np.argmin(e_DFT)]))
    plt.plot(r_MCPDFT, e_MCPDFT-min(e_MCPDFT), 'b', label='MCPDFT ($\Delta E_d$={0:.3f}@{1:.2f}$\AA$)'.format(e_MCPDFT[-1]-min(e_MCPDFT), r_MCPDFT[np.argmin(e_MCPDFT)]))
    plt.xlabel("Interatomic Distance ($\AA$)")
    plt.ylabel("$\Delta$ Potential Energy (eV)")
    plt.title("${0}_2$ Bond Dissociation".format(element))
    plt.legend()
    plt.tight_layout()
```
